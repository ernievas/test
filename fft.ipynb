```python
import numpy as np  # para calcular, numerical python.
import matplotlib.pyplot as plt # para graficar, matplotlib.
# gráficos dentro de la notebook
%matplotlib inline 
```

## Ejemplo 1

El dato es una función constante, $f(x,y)=c$.


```python
nx,ny = 5,5
s = np.ones(nx*ny).reshape(nx,ny)   # constante
```

Graficamos el dato 2D muestreado

```python
plt.close("all")
plt.figure(1,figsize=(5,5))
#plt.imshow(s,origin="lower",interpolation=None)
plt.pcolor(s, edgecolors='k', linewidths=2)
plt.show()
```

Calculamos la transformada discreta de Fourier 2D y graficamos el espectro de amplitud.

```python
S = np.fft.fft2(s)

plt.figure(2,figsize=(5,5))
#plt.imshow(np.abs(S),origin="lower",interpolation=None)
plt.pcolor(np.abs(S), edgecolors='k', linewidths=2)
plt.show()
```

Calculamos la TDF 2D por medio de la TDF 1D. Comparamos con la subrutina *fft2d* de *numpy*

```python
SS = np.zeros_like(s,dtype=complex)

# TDF por fila y por columna
for i in np.arange(nx):
    SS[i,:] = np.fft.fft(s[i,:])
for j in np.arange(ny):
    SS[:,j] = np.fft.fft(SS[:,j])
    
print ("check: ", np.allclose(S,SS))
```

## Ejemplo 2

El dato es una una función cajón o *boxcar* 2D.

```python
nx,ny = 5,5
s = np.pad(s,(5,5),'constant')  
S = np.fft.fft2(s)

plt.figure(3,figsize=(5,5))
#plt.imshow(s,origin="lower",interpolation=None)
plt.pcolor(s, edgecolors='k', linewidths=2)
plt.show()

plt.figure(4,figsize=(5,5))
#plt.imshow(np.abs(S),origin="lower",interpolation=None)
plt.pcolor(np.abs(S), edgecolors='k', linewidths=2)
plt.show()
```
